## 7. Available Columns

List all columns available in the dataset.


```python
#The available columns were listed to understand which variables can be used for player performance analysis, comparison, and future feature engineering.
columns = df.columns.tolist()

print(f"The dataset contains {len(columns)} columns:")
for column in columns:
    print("-", column)
```

## 8. Missing Values

Check if the dataset contains missing values.

```python
# Missing values were checked to identify incomplete columns and understand whether they may affect future analysis, cleaning, feature engineering, or modeling steps.
missing_values = df.isnull().sum()
missing_percentage = (missing_values / len(df)) * 100

missing_summary = pd.DataFrame({
    "missing_values": missing_values,
    "missing_percentage": missing_percentage
})

missing_summary[missing_summary["missing_values"] > 0].sort_values(
    by="missing_values",
    ascending=False
)
```

## 9. Duplicates

Check if the dataset contains duplicated rows.

```python
duplicate_rows = df.duplicated().sum()

print(f"The dataset contains {duplicate_rows} duplicated rows.")
```
