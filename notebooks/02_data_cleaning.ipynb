## 5. Fix Numeric Columns Imported as Text

Some columns that represent numeric values may be imported as text due to formatting issues in the original CSV file.

To fix this, columns that should remain categorical, such as player name, nationality, position, squad, and competition, were separated from the rest of the dataset.

All remaining columns were converted to numeric types using `pd.to_numeric()`. Values that could not be converted were transformed into missing values so they can be reviewed later.

```python
# Check data types of columns to identify which ones are strings and which ones are numeric. This will help in the next steps of data cleaning and analysis.
column_types = dict(df.dtypes)

string_columns = [col for col, dtype in column_types.items() if dtype == "str"]

numeric_columns = [col for col, dtype in column_types.items() if col not in string_columns]

for col in numeric_columns:
    df[col] = pd.to_numeric(df[col], errors="coerce")

df.info()
```

```python
# After converting the columns to numeric, we can check how many missing values were introduced due to conversion errors. This will help us understand the data quality and decide on further cleaning steps.
missing_after = df.isna().sum()
new_missing = missing_after - missing_before
missing_report = pd.DataFrame({
    "missing_before": missing_before,
    "missing_after": missing_after,
    "new_missing": new_missing
})

print(missing_report)
```

```python
# Save the cleaned dataset to a new CSV file
df.to_csv("../data/processed/players_clean.csv", index=False)
```

# 6. Remove players with less than 300 minutes

Players with fewer than 300 minutes will be removed from the analysis dataset to reduce noise caused by very small sample sizes.

Players with limited playing time may have inflated or unstable per-90 metrics, making comparisons unfair against players who played a more representative number of minutes.

The original raw dataset will remain unchanged. This filter will only be applied to the cleaned dataset used for analysis.

```python
df_filtered = df[df["minutes"] >= 300]
print("Shape of original dataset:", df.shape)
print("Shape of filtered dataset:", df_filtered.shape)

# Save the cleaned dataset to a new CSV file
df_filtered.to_csv("../data/processed/players_clean_min300.csv", index=False)
```
